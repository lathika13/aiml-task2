```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x="Pclass", y="Age", data=df)
plt.title("Age vs Pclass")
plt.show()
```

```python
sns.countplot(x="Survived", hue="Sex", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(
    df[["Age","Fare","SibSp","Parch","Survived"]].dropna(),
    hue="Survived"
)
plt.show()
```

Findings:

1. Females have a higher survival rate than males.
2. 1st class passengers show higher survival probability.
3. Age has missing values and needs imputation.
4. Fare is strongly related to passenger class.
5. Cabin column is mostly empty and not usable in raw form.
6. Low correlation between many variables  no multicollinearity.
